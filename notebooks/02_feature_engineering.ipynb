**FEATURE ENGINEERING **

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load raw data
df = pd.read_csv(r"D:\final_project\data\tabular_data\tabular data.csv")


# Drop patient_id (not useful for ML)

df = df.drop(columns=['patient_id'])


# Handle Missing Values

df.fillna(df.median(numeric_only=True), inplace=True)


# Encode categorical columns

df = pd.get_dummies(df, columns=['gender'], drop_first=True)


# Separate features & targets

X = df.drop(columns=['outcome', 'length_of_stay_days'])
y_class = df['outcome']
y_reg = df['length_of_stay_days']


# Feature Scaling

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

clean_df = pd.DataFrame(X_scaled, columns=X.columns)
clean_df['outcome'] = y_class.values
clean_df['length_of_stay_days'] = y_reg.values


# Save Clean Data

clean_df.to_csv(r"D:\final_project\data\processed\clean_healthcare.csv", index=False)

print("Clean dataset saved!")
```
